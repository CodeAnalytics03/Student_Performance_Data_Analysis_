```python
import pandas as pd
data = pd.read_csv("students.csv")
print(data)
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
data["Average"] = data['Math'] + data['Science'] + data['English'] / 3
print(data)
```

```python
top_student = data.loc[data["Average"].idxmax()]
print(top_student)
```

```python
print("Average Math Score: ", data['Math'].mean())
```

```python
print("Highest Science Score: ", data["Science"].max())
```

```python
print("Lowest English Score: ", data['English'].min())
```

```python
correlation = data['StudyHours'].corr(data['Average'])
print("Correaltion Between Study Hours and Average is ", correlation)
```

```python
import matplotlib.pyplot as plt
plt.bar(data["Student"], data["Average"])
plt.title("Student Performance")
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.show()
```
